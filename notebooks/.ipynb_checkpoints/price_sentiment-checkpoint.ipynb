```python
import pandas as pd
import requests

# Load the dataset
file_path = "../artifacts/lat_log_cuision.csv"
df = pd.read_csv(file_path)

# Function to convert currency to USD using real-time exchange rates
def convert_to_usd(price, currency):
    try:
        price = float(price)  # Ensure price is a float
        if currency == "USD":
            return price  # Already in USD, no conversion needed
        
        # Fetch real-time exchange rate
        response = requests.get(f"https://v6.exchangerate-api.com/v6/ef480af61c330493a658eebe/latest/USD{currency}")
        data = response.json()
        
        # Get conversion rate
        exchange_rate = data["rates"].get("USD", None)
        if exchange_rate is not None:
            return price / exchange_rate  # Convert to USD
        else:
            return None  # Handle cases where conversion is not possible
    except:
        return None  # Handle errors (e.g., missing values or invalid currency codes)

# Apply the conversion function
df["Price in USD"] = df.apply(lambda row: convert_to_usd(row["Average Cost for two"], row["Currency"]), axis=1)

# Check the updated data
print(df[["Average Cost for two", "Currency", "Price in USD"]].head())
```

```python
import pandas as pd

# Load dataset (original remains unchanged)
file_path = "../artifacts/lat_log_cuision.csv"
df = pd.read_csv(file_path)

# Correct currency labels
currency_mapping = {
    'Botswana Pula(P)': 'BWP',
    'Brazilian Real(R$)': 'BRL',
    'Dollar($)': 'USD',
    'Emirati Diram(AED)': 'AED',
    'Indian Rupees(Rs.)': 'INR',
    'Indonesian Rupiah(IDR)': 'IDR',
    'NewZealand($)': 'NZD',
    'Pounds(��)': 'GBP',  # Check encoding issue here
    'Qatari Rial(QR)': 'QAR',
    'Rand(R)': 'ZAR',
    'Sri Lankan Rupee(LKR)': 'LKR',
    'Turkish Lira(TL)': 'TRY'
}

# Apply currency mapping
df["Currency"] = df["Currency"].map(currency_mapping).fillna(df["Currency"])

# Exchange rate dictionary
exchange_rates = {
    'USD': 1, 'AED': 3.6725, 'BWP': 13.8658, 'BRL': 5.836, 'GBP': 0.794,
    'INR': 87.4519, 'IDR': 16571.5137, 'NZD': 1.7876, 'QAR': 3.64, 'ZAR': 18.5891,
    'LKR': 294.9457, 'TRY': 36.5357
}

# Function to convert prices to USD
def convert_to_usd(price, currency):
    if currency in exchange_rates:
        return round(price / exchange_rates[currency], 2)  # Convert and round to 2 decimals
    return price  # Keep original price if currency is missing

# Convert prices and store in a new column
df["Price_in_USD"] = df.apply(lambda row: convert_to_usd(row["Average Cost for two"], row["Currency"]), axis=1)

# Save the processed data as a new file
cleaned_file_path = "../artifacts/cleaned_data.csv"
df.to_csv(cleaned_file_path, index=False)

# Show updated results
print(df[["Average Cost for two", "Currency", "Price_in_USD"]].head())
```

```python
import pandas as pd

# Load the cleaned dataset (after currency conversion)
cleaned_file_path = "../artifacts/cleaned_data.csv"
df = pd.read_csv(cleaned_file_path)

### STEP 1: REMOVE OUTLIERS USING IQR ###
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)  # 25th percentile
    Q3 = df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile Range

    # Define outlier boundaries
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter data within the valid range
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df

# Apply the function to remove outliers
df_no_outliers = remove_outliers(df, "Price_in_USD")

# Save the dataset without outliers
outliers_removed_path = "../artifacts/outliers_removed.csv"
df_no_outliers.to_csv(outliers_removed_path, index=False)

# Print summary
print("Original Dataset:", df.shape)
print("After Removing Outliers:", df_no_outliers.shape)

# Show a preview of the cleaned dataset
print(df_no_outliers[["Price_in_USD"]].head())
```
